# Matplotlib

```python
%matplotlib inline
from matplotlib import pyplot as plt, style
```

### Plotting a line on the chart

```python
plt.plot([1, 2, 3], [2, 5, 7])
# plt.show()
```

### Giving Heading to the chart and Labels to the axes

```python
plt.plot([1, 2, 3], [2, 5, 7])
plt.title("My Data")
plt.ylabel("Ordinate")
plt.xlabel("Abscissa")
# plt.show()
```

### Plotting two lines

```python
x1 = [1, 5, 4]
y1 = [23, 45, 89]
x2 = [3, 7, 3]
y2 = [5, 8, 9]

plt.plot(x1, y1)
plt.plot(x2, y2)
```

### Changing the style of the plot, plot functions, and using legends for different lines

```python
x1 = [1, 5, 4]
y1 = [23, 45, 89]
x2 = [3, 7, 3]
y2 = [5, 8, 9]

style.use("ggplot")
plt.plot(x1, y1, label="line1", linewidth=4)
plt.plot(x2, y2, label="line2", alpha=0.5)
plt.legend()
```

### Scatter Plot

```python
x1 = [1, 5, 4]
y1 = [23, 45, 89]
x2 = [3, 7, 3]
y2 = [5, 8, 9]

style.use("ggplot")
plt.scatter(x1, y1, label="line1", linewidth=4)
plt.scatter(x2, y2, label="line2", alpha=0.5)
plt.legend()
```

### Histogram

```python
num_bins = 5
plt.hist([21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100], num_bins, facecolor='blue', alpha=0.5)
```
